```python
#necessary libraries
import pandas as pd
import itertools
import openpyxl
import pprint
```

```python
#importing data
df = pd.read_excel('グレイ単語カード ひらがな無.xlsx')
```

```python
begin_index = df[df['カードNo.']==1].index[0]
end_index = df[df['カードNo.']==2].index[0]
df.iloc[begin_index:end_index-1,:]
```

```python
#setting year
year = 2020

#last row of the excel
end_index_num = 4542

#setting test ranges
test1 = list(itertools.chain(range(1,23),range(35,40),range(115,118),range(135,138),range(140,142),range(163,164),range(165,166),range(176,178),range(179,180),range(193,197),range(198,200),range(201,202),range(213,216),range(227,234)))
test2 = list(itertools.chain(range(75,76),range(142,150),range(202,205),range(260,270),range(285,286),range(307,308),range(320,325),range(337,340),range(347,349),range(352,355),range(366,367)))
test3 = list(itertools.chain(range(23,29),range(40,55),range(59,60),range(61,63),range(64,74),range(76,92),range(97,103),range(150,153),range(156,159),range(197,198),range(208,213),range(340,347),range(349,352)))
test4 = list(itertools.chain(range(29,35),range(55,59),range(60,61),range(63,64),range(74,75),range(92,97),range(103,115),range(128,133),range(138,140),range(153,155),range(200,201),range(205,208),range(270,280),range(380,394)))
test5 = list(itertools.chain(range(118,128),range(133,135),range(155,156),range(159,163),range(164,165),range(166,176),range(178,179),range(180,193),range(280,285),range(301,309),range(328,334),range(336,337),range(355,372)))
test6 = list(itertools.chain(range(216,227),range(234,260),range(286,301),range(309,328),range(334,336),range(372,380)))
tests = [test1, test2, test3, test4, test5, test6]

#color format
color = [openpyxl.styles.PatternFill(patternType='solid', fgColor='FF0000', bgColor='FF0000')] * 373
```

```python
test_num = 1
for each_test in tests:
    test_path = "{}_test{}.xlsx".format(year, test_num)
    excel_writer = pd.ExcelWriter(test_path)
    content = pd.DataFrame({})
    for section_number in each_test:
        begin_index = df[df['カードNo.']==section_number].index[0]
        if section_number != 393:
            end_index = df[df['カードNo.']==section_number+1].index[0]
        else:
            end_index = end_index_num
        content = pd.concat([content, df.iloc[begin_index:end_index-1,:]])
    test_num += 1
    content.set_index('カードNo.', inplace=True)
    content.to_excel(excel_writer, 'Sheet1')
    excel_writer.save()
```

```python
#coloring cells
test_num = 1

for each_test in tests:
    test_path = "{}_test{}.xlsx".format(year, test_num)
    excel_writer = openpyxl.load_workbook(test_path)
    sheet = excel_writer['Sheet1']
    df2 = pd.read_excel(test_path)
    for section_number in range(len(each_test)):
        begin_index = df2[df2['カードNo.']==each_test[section_number]].index[0] + 2
        if section_number < len(each_test) - 1:
            end_index = df2[df2['カードNo.']==each_test[section_number+1]].index[0]
        else:
            end_index = len(df2)+1
        for i in range(begin_index, end_index+1):
            cell = 'A{}'.format(i)
            sheet[cell].fill = color[section_number]
    excel_writer.save(test_path)
    test_num += 1
```
